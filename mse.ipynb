```python
def point_mse(arr, gt):
    if arr[0]==-1 or gt[0]==-1:
        return 'Nan'
    else:
        return np.sqrt((arr[0]-gt[0])**2 + (arr[1]-gt[1])**2)
```

```python
mse_array = []
points_name = []

for i in range(2300):
        mse_array.append([point_mse(points_label_2[i], points_pred_2[i])])

for j in range(46):
        points_name.append(ID[j])

points_name_2 = points_name * 50
```

```python
len(mse_array)
```

```python
len(points_name_2)
```

```python
mse_df = pd.DataFrame(mse_array, columns=['mse'])
points_name_df = pd.DataFrame(points_name_2, columns=['Landmark name'])

mse_df_2 = pd.concat([points_name_df, mse_df], axis = 1)
```

```python
mse_df_2
```

```python
mse_df_2[mse_df_2['Landmark name'] == ID[0]].reset_index()['mse']
```

```python
df_list = []

for i in range(46):
    df_list.append(mse_df_2[mse_df_2['Landmark name'] == ID[i]].reset_index()['mse'])

df_all = pd.concat(df_list, axis=1, ignore_index=True, names=[ID])

df_all.columns=ID
```
